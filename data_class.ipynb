```python
from dataclasses import dataclass
```

```python
@dataclass()
class Student:
    name: str
    age: int
```

```python
stu = Student("sidd", 22)
print(stu)
```

```python
@dataclass(frozen=True)#isse ye class immutable hojati hai
class Car:
    color: str
    model: str
    brand: str
```

```python
my_car = Car("black", "M5", "BMW")
print(my_car)
```
